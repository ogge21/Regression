```python
df['numberOfRooms']=df['numberOfRooms']**2
```

```python
df["cityCode"]=df["cityCode"].astype("object")
```

```python
df["age"]=2022-df["made"]
```

```python
abs(df.corr()['price'].sort_values(ascending=False))
```

```python
x=df.drop(['price','id'],axis=1)
y=df[['price']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.metrics import r2_score, mean_squared_error
```

```python
x
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
model=lr.fit(x_train,y_train)
```

```python
tahmin=model.predict(x_test)
```

```python
(mean_squared_error(tahmin,y_test))**0.5
```

```python
r2_score(tahmin,y_test)
```

```python
from sklearn.linear_model import Ridge, Lasso
R=Ridge()
L=Lasso()
```

```python
R.fit(x_train,y_train) 
```

```python
tahmin2=R.predict(x_test)
```

```python
(mean_squared_error(tahmin2,y_test))**0.5
```

```python
r2_score(tahmin2,y_test)
```

```python
L.fit(x_train,y_train)
```

```python
tahmin3=L.predict(x_test)
```

```python
(mean_squared_error(tahmin3,y_test))**0.5
```

```python
r2_score(tahmin3,y_test)
```
